```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)  #(시작,끝,갯수) 0에서 10까지 100개의 균일한 점을 찍어라
```

```python
x
```

```python
np.sin(x)
```

```python
plt.plot(x, np.sin(x))
```

```python
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
# 이미지 저장하기
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
fig.savefig('my_figure.png')  #확장자까지 지정해주는 편이 좋음
```

```python
#한 이미지 안에 각각의 그래프 출력
plt.subplot(2,1,1)   #가로로 봤을땐 2개, 세로로 봤을땐 1개, 그중에서 첫번째
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
```

```python
# 선의 색깔
plt.plot(x, np.sin(x-0), color='blue')
plt.plot(x, np.sin(x-1), color='g')   #단축색깔(b는 파랑, k는 검정)
plt.plot(x, np.sin(x-2), color='0.75')   #음영(1은 흰색 검정색)
plt.plot(x, np.sin(x-3), color='#FFDD44')   #RGB
plt.plot(x, np.sin(x-4), color=(1.0,0.2,0.3))   #RGB
plt.plot(x, np.sin(x-5), color='chartreuse')
```

```python
# 선의 형태
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

```python
plt.plot(x, x+0, '-g')
plt.plot(x, x+1, '--c')
plt.plot(x, x+2, '-.k')
plt.plot(x, x+3, ':r')
```
